### Exercice 2

#### a)

```python
def G(X):
    return numerical_integral(((1-exp(-x))/(1 + x**2)), 0, X)[0]
```

```python
plot(((1-exp(-x))/(1 + x**2)), 0,10, color = 'blue') + plot(G, 0, 10, color = 'red') 
```

```python
X = tan((pi / 2) - 10**-3)
show("L'integrale imropre tend vers :", numerical_approx(G(X), digits = 4))
```

#### b)

```python
def H(X):
    return numerical_integral(exp(-x**2), 0, X)[0]
```

```python
plot(exp(-x**2), 0,10, color = 'blue') + plot(H, 0, 10, color = 'red') 
```

```python
X = sqrt(-ln((2 / sqrt(pi)) * 10**-3))
show("L'integrale imropre tend vers :", numerical_approx(H(X), digits = 4))
```
