### Exercice 2

```python
L = []
for n in range(100, 10000, 100):
    somme = 0
    for i in range(n):
        x = (random()*4)-2
        y = (random()*4)-2
        if ((((0.05 + x)**2 + y**2 + 1)**2) <= 4 * x**2 + 1.4):
            somme += (((0.05 + x)**2 + y**2 + 1)**2 <= 4 * x ** 2 + 1.4)
    L.append(somme * 16 / n)
list_plot(L)
```

```python
print("Comparé à la première méthode, la précision est meilleur et le résultat est plus correct")
```
