```python
x = var('x')
y = var('y')
xR = 0
yR = 0
pointIn = 0
aire = 0
fil = [  [(-2,-2), (0,-2), (-2,0), (0,0)],  [(0,-2), (2,-2), (0,0), (2,0)],  [(-2,0), (0,0), (-2,1), (0,1)],  [(0,0), (2,0), (0,1), (2,1)]  ] 

def LargeurHauteur(pt): #renvoie la Largeur et la Hauteur d'un rectangle
    x0, y0 = pt[0]
    x1, y1 = pt[1]
    x2, y2 = pt[2]
    return (-x0 + x1), (-y0 + y2) #largeur, hauteur


def aireRect(pt): #renvoie l'aire d'un rectangle
    largeur, hauteur = LargeurHauteur(pt)
    return largeur * hauteur


for i in range(0, 20000): 
    pointIn = 0
    for j in range(0, 4):
        xR, yR = fil[0][j]
        if (((0.05 + xR)**2 + yR**2 + 1)**2 - (4*xR**2 + 1.4) <= 0): #test si un point est dans la cacahuète
            pointIn += 1

    if(pointIn == 4): #si le rectangle est dans la cacahuète
        aire += aireRect(fil[0])
    elif(pointIn > 0): #si le rectangle n'est pas entierement dans la cacahuète
        LR, HR = LargeurHauteur(fil[0]) #on récupère la largeur et la hauteur qu'on divisera ensuite par 2 (pour les sous-rectangles)
        LR /= 2
        HR /= 2
        xR, yR = fil[0][0] 
        fil.append([ (xR,    yR),    (xR+LR,   yR),    (xR,    yR+HR),   (xR+LR,   yR+HR) ])   #bas/gauche
        fil.append([ (xR+LR, yR),    (xR+LR*2, yR),    (xR+LR, yR+HR),   (xR+LR*2, yR+HR) ])   #bas/droite
        fil.append([ (xR,    yR+HR), (xR+LR,   yR+HR), (xR,    yR+HR*2), (xR+LR,   yR+HR*2) ]) #haut/gauche
        fil.append([ (xR+LR, yR+HR), (xR+LR*2, yR+HR), (xR+LR, yR+HR*2), (xR+LR*2, yR+HR*2) ]) #haut/droite
    #si le rectangle se situe à l'extérieur on ne le redécoupe pas
    fil.pop(0) #on retire le dernier rectangle

show(numerical_approx(aire, digits =4))
show("l'aire via la methode quadtree tend aussi vers 2.9")
```
