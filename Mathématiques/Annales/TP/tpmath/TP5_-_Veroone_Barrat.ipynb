```python
f(x, y) = (x**2 * exp(2*x-1) + 4*y**2) * exp(-x**2 - y**2)
f(1, 1)
```

```python
def methode_gradient(P0, h, N, tol_p, tol_f):
    Pn = P0
    n = 0
    
    if (tol_f < 0):
        return 0
    
    Pn2 = (Pn[0] + h * f.gradient()(Pn[0], Pn[1])[0], Pn[1] + h * f.gradient()(Pn[0], Pn[1])[1]) 
    
    while (n < N and ((sqrt((Pn2[0] - Pn[0])**2 + (Pn2[1] - Pn2[0])**2)) < tol_p) and (f(Pn2[0], Pn2[1]) - f(Pn[0],Pn2[1]) < tol_f)):
        Pn2 = (Pn[0] + h * f.gradient()(Pn[0], Pn[1])[0], Pn[1] + h * f.gradient()(Pn[0], Pn[1])[1]) 
        Pn = Pn2
        n += h
    
    return(Pn2)
```

```python
P = (1, 2)
```

```python
f.gradient()(P[0], P[1])
```

```python
result = methode_gradient(P, 3, 10000, 0.001, 0.001)
show("x : ", numerical_approx(result[0]), ", y : ", numerical_approx(result[1]))
```
