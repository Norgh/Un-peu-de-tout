```python
def is_reflexive (liste):
    reflexive = true
    for i in range(len(liste)):
        if (liste[i][i] != 1):
            reflexive = false
    return reflexive
```

```python
def is_symetrique (liste):
    for i in range(len(liste)):
        for j in range(i, len(liste)):
            if (liste[i][j] != liste[j][i]):
                return false
    return true
```

```python
def is_antisymetrique (liste):
    antisymetrique = true
    for i in range(len(liste)):
        for j in range(len(liste)):
            if (liste[i][j] == 1 and liste[i][j] == liste[j][i] and i != j):
                antisymetrique = false
                break 
        if (not(antisymetrique)):
            break
    return antisymetrique
```

```python
def is_transitive (liste):
    transitive = true
    for i in range(len(liste)):
        for j in range(len(liste)):
            if(liste[i][j]==1):
                for k in range(len(liste)):
                    if (liste[j][k]==1 and liste[i][k] != 1):
                        transitive = false
    return transitive
```

```python
def is_M_reflexive (m):
    reflexive = true
    for i in range(m.nrows()):
        if (m[i][i] != 1):
            reflexive = false
    return reflexive
```

```python
def is_M_symetrique (m):
    for i in range(m.nrows()):
        for j in range(i, m.ncols()):
            if (m[i][j] != m[j][i]):
                return false
    return true
```

```python
def is_M_antisymetrique (m):
    antisymetrique = true
    for i in range(m.nrows()):
        for j in range(m.ncols()):
            if (m[i][j] == 1 and m[i][j] == m[j][i] and i != j):
                antisymetrique = false
                break 
        if (not(antisymetrique)):
            break
    return antisymetrique
```

```python
def is_M_transitive (m):
    for i in range(m.nrows()):
        for j in range(m.ncols()):
            if(m[i][j]==1):
                for k in range(m.ncols()):
                    if (m[j][k]==1 and m[i][k] != 1):
                        return false
    return true
```

```python
def test_M_relation (m):
    print("La relation est :")
    if (is_M_reflexive(m)):
        print("réfléxive ")
    if (is_M_symetrique(m)):
        print("symétrique ")
    if (is_M_antisymetrique(m)):
        print("antisymétrique ")
    if (is_M_transitive(m)):
        print("transitive ")
```
