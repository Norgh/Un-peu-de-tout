```python
def F(X):
    return numerical_integral(sin(x)/x, 0, X)[0]

# EXERCICE 1
print('a)')
show(plot(sin(x)/x, 0, 100, color = 'chartreuse') + plot(F, 0, 100, color = 'darkkhaki'))
print("L'intégrale impropre semble converger en 1.6 environ, ce qui semble être pi/2, lorsque x tend vers +l'infini")
```

```python
x=var('x')
X=(2*(1/0.001)-pi)
show(numerical_approx(F(X)))
print("On retrouve bien une valeur proche de 1.6, c'est à dire de pi/2")
```

```python
#EXERCICE 2
print("a)")
def G(X):
    return numerical_integral(((1-exp(-x))/(1 + x*x)), 0, X)[0]

show(plot(((1-exp(-x))/(1 + x*x)), 0, 50, color = 'sienna') + plot(G, 0, 50, color = 'salmon')) 
```

```python
def G(X):
    return numerical_integral(((1-exp(-x))/(1 + x*x)), 0, X)[0]

X=tan((pi/2)-10^-3)
print("G tend vers :")
print(numerical_approx(G(X)))
```

```python
print("b)")
def H(X):
    return numerical_integral(exp(-x*x), 0, X)[0]

plot(exp(-x*x), 0, 5, color = 'burlywood') + plot(H, 0, 5, color = 'chocolate') 
```

```python
def H(X):
    return numerical_integral(exp((-x*x)), 0, X)[0]
X=sqrt(-ln((2/sqrt(pi))*10^-3))
print("H tend vers :")
print(numerical_approx(H(X)))
```
