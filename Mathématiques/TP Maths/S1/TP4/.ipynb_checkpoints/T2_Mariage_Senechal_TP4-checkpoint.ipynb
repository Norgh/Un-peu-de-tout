```python
f(z) = z^2 - z - 1
complex_plot(f, (-2,2), (-2,2), aspect_ratio = 1)
```

On observe la position de 2 pôles simples (les arguments font un tour dans le sens anti-horaire) en z vaut environ 1,6 et -0,6. 
En fait il s'agit de ((1 + sqrt(5)) / 2) et ((1 - sqrt(5)) / 2). 
On peut remarquer que lorsqu'on s'éloigne, on ne fait plus la différence avec z² (zéro double à l'origine, les arguments font 2 tours complets). 
On voit que |f(z)| tend vers l'infini quand on s'éloigne.

```python
complex_plot(f, (-100,100), (-100,100), aspect_ratio = 1)
```

```python
g(z) = (1/z) + z 
complex_plot(g, (-2,2), (-2,2), aspect_ratio = 1)
```

g(z) = (1+ z²) / z : on voit bien les zéros ( points noirs ) en z = i et z = -i et le pôle simple en z = 0 ( point de couleur blanche autour duquel les arguments tournent une fois dans le sens horaire).

```python
h(z) = cos(z) 
complex_plot(h, (-3*pi,3*pi), (-3,3), aspect_ratio = 1)
```

On voit bien la (2*pi)-périodicité dans la direction réelle (dans la direction imaginaire, les modules tendent vers l'infini) et les zéros simples en tous les (pi/2) + k*pi. Sur l'axe réel on voit bien les alternances de valeurs positives (rouges) et négatives (cyan) en passant par les zéros (noirs).

Exercice 2:

```python
f2(z) = (z^2 + z + 1) / (2 - z)
def fN(z, N):
    return 1/2 + 3/4*z + 7/2*sum( z^n/2^n , n, 2, N)
```

```python
complex_plot(f2, (-5,5), (-5,5), aspect_ratio = 1)
```

```python
complex_plot(fN(z,100), (-5,5), (-5,5), aspect_ratio = 1, plot_points = 500)
```

En augmentant N, on peut voir que : 
- Les sommes partielles convergent bien vers les valeurs de f(z) au voisinage de z = 0
- on a un disque de rayon R = 2 (rayon de convergence)

```python
g2(z) = 1 / z
def gN(z, N):
    return sum( (-1)^n / 3^(n+1)*(z-3)^n , n, 0, N)
```

```python
complex_plot(g2, (-6,6), (-6,6), aspect_ratio = 1)
```

```python
complex_plot(gN(z,30), (-6,6), (-6,6), aspect_ratio = 1, plot_points = 500)
```

En augmentant N, on peut voir que : 
- Les sommes partielles convergent bien vers les valeurs de f(z) au voisinage de z = 3
- on a un disque de rayon R = 3 (rayon de convergence)

```python
h2(z) = atan(z)
def hN(z, N):
    return sum( (-1)^n / (2*n+1)*z^(2*n+1) , n, 0, N)
```

```python
complex_plot(h2, (-2,2), (-2,2), aspect_ratio = 1)
```

```python
complex_plot(hN(z,40), (-2,2), (-2,2), aspect_ratio = 1, plot_points = 1000)
```

On a ici R = 1 (rayon de convergence)
