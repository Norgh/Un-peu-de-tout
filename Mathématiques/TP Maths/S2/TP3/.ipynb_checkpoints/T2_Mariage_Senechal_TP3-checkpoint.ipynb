## Exercice 1

Générer toutes les relations sur $E_n = [\![0,n[\![$, _i.e._ toutes les parties de $E_n \times E_n$. Il y a en général $2^{2n} = 4^n$, on peut réutiliser ce qui a été fait dans le TP1.

```python
def parties(E):
    
    if E == []:
    
        return [[]]
        
    else:
        
        x, EE = E[0], E[1:]
        tmp = parties(EE)
        
        return [ [x] + A for A in tmp ] + tmp
    
    
def all_relations(n):
    
    return parties([(x,y) for x in range(n) for y in range(n)])
    
```

```python
all_relations(3)
```

```python
len(all_relations(3)) # 2^9 ok
```

On pourrait aussi alternativement générer toutes les matrices binaires...

```python
def mat_adj(n, rel):
 
    A = matrix(n)
    
    for x in range(n):
        for y in range(n):
            if(x,y) in rel:
                A[x,y] = 1
    return A

def all_matrices(n):
    
    return [mat_adj(n, R) for R in all_relations(n)]
    
```

Vérification des propriétés (on peut le faire sur les listes ou sur les matrices).
Ici sur les matrices on n'a pas besoin à côté de connaître la taille de l'ensemble.

```python
def cap(A):
    n=A.nrows()
    res = matrix(n)
    for x in range(n):
        for y in range(n):
            if A[x,y] >= 1:
                res[x,y] = 1
            if A[x,y] <= 0:
                res[x,y] = 0
    return res

def oplus(A,B):
    return cap(A+B)

def otimes(A,B):
    return cap(A*B)

def is_contained(A,B):
    return oplus(A,B)==B
```

```python
def is_reflexive(A):
    return is_contained(1,A)

all_refl = [ A for A in all_matrices(3) if is_reflexive(A)]
all_refl  # 2^6 ok
```

```python
def is_symmetric(A):
    
    return is_contained(A.transpose(),A)

all_sym = [ A for A in all_matrices(3) if is_symmetric(A)]
all_sym  # 2^6 ok
```

```python
def is_antisymmetric(A):  ## old
    
    B = A + A.transpose()    # les 2 correspondent aux paires xRy et yRx
    
    C = B - cap(B)           # les 1 correspondent aux paires xRy et yRx
    
    return is_cntained(C,1)  # on veut seulement des 1 sur la diagonale

def is_antisymmetric(A):
    
    return is_contained(otimes(A,A), 1)

all_sym = [ A for A in all_matrices(3) if is_antisymmetric(A)]
all_sym # 2^3 * 3^3 ok
```

```python
def is_transitive(A):
    
    return is_contained(otimes(A,A), A)

all_sym = [ A for A in all_matrices(3) if is_transitive(A)]
all_sym
```
